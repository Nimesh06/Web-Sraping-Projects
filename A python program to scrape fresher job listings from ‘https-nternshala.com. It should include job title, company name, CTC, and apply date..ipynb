# Write a python program to scrape fresher job listings from ‘https://internshala.com/’. It should include job title, company name, CTC, and apply date.

```python
# importing all the required libraries
from bs4 import BeautifulSoup
from selenium import webdriver
import requests
import pandas as pd

# sending request to the web page to get the required information
page=requests.get("https://internshala.com/fresher-jobs")                     # book, author, genere, review
page
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content)
```

```python
# extracting all the header tags
# fetching all the header tiles from the website
jobtitle=soup.find_all('div',class_="heading_4_5 profile")
jobtitle
```

```python
compname=soup.find_all('div',class_="heading_6 company_name")
compname
```

```python
ctc=soup.find_all('div',class_="item_body")
ctc
```

```python
date=soup.find_all('div',class_="item_body")
date
```

```python
job_title=[]
for i in jobtitle:
    job_title.append(i.text.replace('\n',''))
job_title
```

```python
company_name=[]
for i in compname:
    company_name.append(i.text.strip())  #company_name.append(i.text.replace('\n',''))
company_name
```

```python
ctc_range=[]
for i in ctc:
    ctc_range.append(i.text.strip())
ctc_range=ctc_range[1::3]
ctc_range
```

```python
apply_date=[]
for i in date:
    apply_date.append(i.text.strip())
apply_date=apply_date[2::3]
apply_date
```

```python
print(len(job_title),len(company_name),len(ctc_range),len(apply_date))
```

```python
joblist=pd.DataFrame({})
joblist['Job Title']=job_title
joblist['Company Name']=company_name
joblist['Ctc Range']=ctc_range
joblist['Apply Date']=apply_date
joblist
```
