```python
soup=BeautifulSoup(page.content)
```

```python
# extracting all the header tags
# fetching all the header tiles from the website
housetitle=soup.find_all('h2',class_="heading-6 font-semi-bold nb__1AShY")
housetitle
```

```python
location=soup.find_all('div',class_="nb__2CMjv")
location
```

```python
area=soup.find_all('div',class_="nb__3oNyC")
area
```

```python
emi=soup.find_all('div',class_="font-semi-bold heading-6")
emi
```

```python
price=soup.find_all('div',class_="font-semi-bold heading-6")
price
```

```python
house_title=[]
for i in housetitle:
    house_title.append(i.text)
house_title
```

```python
location_name=[]
for i in location:
    location_name.append(i.text)
location_name
```

```python
area_sqft=[]
for i in area:
    area_sqft.append(i.text)
area_sqft
```

```python
emi_month=[]
for i in emi:
    emi_month.append(i.text)
emi_month=emi_month[1::3]
emi_month
```

```python
price_total=[]
for i in price:
    price_total.append(i.text)
price_total=price_total[2::3]
price_total
```
